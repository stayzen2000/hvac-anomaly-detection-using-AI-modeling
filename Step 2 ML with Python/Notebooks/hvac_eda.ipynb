```python
import pandas as pd

# CSV file is being read first

df = pd.read_csv("../../Step 1 Cleaning with SQL/hvac_clean.csv") 

df.head()
```

```python
# This block performs an initial data audit by checking the datasetâ€™s size, class distribution, 
# and presence of missing values to ensure the data is clean and ready for modeling.

df.shape
df['anomaly_flag'].value_counts()
df.isna().sum()
```

```python
# Timestamp is made sure it is parsed correctly
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Sorting by timestamp
df = df.sort_values('timestamp')

# Calculating split indices
n = len(df)
train_end = int(0.70 * n)
val_end = int(0.85 * n)

# Creating time-based splits
train_df = df[:train_end]
val_df   = df[train_end:val_end]
test_df  = df[val_end:]

# Printing shapes
len(train_df), len(val_df), len(test_df)
```
