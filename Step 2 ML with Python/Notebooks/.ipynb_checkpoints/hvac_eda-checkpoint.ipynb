```python
import pandas as pd

# CSV file is being read first

df = pd.read_csv("../../Step 1 Cleaning with SQL/hvac_clean.csv") 

df.head()
```

```python
# This block performs an initial data audit by checking the dataset’s size, class distribution, 
# and presence of missing values to ensure the data is clean and ready for modeling.

df.shape
df['anomaly_flag'].value_counts()
df.isna().sum()
```

```python
# Timestamp is made sure it is parsed correctly
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Sorting by timestamp
df = df.sort_values('timestamp')

# Calculating split indices
n = len(df)
train_end = int(0.70 * n)
val_end = int(0.85 * n)

# Creating time-based splits
train_df = df[:train_end]
val_df   = df[train_end:val_end]
test_df  = df[val_end:]

# Printing shapes
len(train_df), len(val_df), len(test_df)
```

```python
# Select features and target
target = 'anomaly_flag'
features = [col for col in df.columns if col not in ['timestamp', 'anomaly_flag', 'system_state']]

# To train the model
X_train = train_df[features]
y_train = train_df[target]

# To evaluate for any fine tuning if needed
X_val = val_df[features]
y_val = val_df[target]

# To test the values
X_test = test_df[features]
y_test = test_df[target]

X_train.shape, X_val.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Imported and created RandomForestClassifier by defining it with features 
rf = RandomForestClassifier(
    n_estimators=200,         # 200 decision trees
    max_depth=None,           # Placing no limit on how deep each tree can grow
    random_state=42,          # Ensuring good reproducibility
    n_jobs=-1                 # Using all CPU cores for training
)

rf.fit(X_train, y_train)      # Model will look at the tranining data to learn the relationship

# Evaluate on validation set
y_pred = rf.predict(X_val)

print(classification_report(y_val, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion matrix on validation set
cm = confusion_matrix(y_val, y_pred)                                            # Takes the true labels (y_val) and predicted labels (y_pred) and generates a 2×2 table
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf.classes_)  # Converts that raw table into a clean visual chart

plt.figure(figsize=(4, 4))
disp.plot(values_format='d')
plt.title("Random Forest - Validation Confusion Matrix")
plt.show()   # Chart display

# Evaluate final model on test set
y_test_pred = rf.predict(X_test)

print("=== Test Set Performance ===")
print(classification_report(y_test, y_test_pred))  # Peints the precision, recall, F1-score, and support only on the test set
```

```python
import numpy as np

# Gets feature importances from the trained model
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

# How many top features to show
top_k = min(15, len(importances))
top_features = [features[i] for i in indices[:top_k]]
top_importances = importances[indices[:top_k]]

plt.figure(figsize=(8, 5))
plt.barh(range(top_k), top_importances[::-1])
plt.yticks(range(top_k), top_features[::-1])
plt.xlabel("Feature importance")
plt.title("Top Most Important Features - Random Forest")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# User-friendly labels for each HVAC feature

friendly_names = {
    "power_kw": "Power Usage (kW)",
    "delta_supply_return": "Temperature Difference: Supply Air - Return Air (°C)",
    "t_return": "Return Air Temperature (°C)",
    "energy_kwh": "Total Energy Used Today (kWh)",
    "t_supply": "Supply Air Temperature (°C)",
    "rh_outdoor": "Outdoor Humidity (%)",
    "temp_error": "Temperature Error (Return - Supply) (°C)",
    "rh_return": "Indoor Humidity (%)",
    "sp_return": "Return Air Setpoint (°C)",
    "rh_supply": "Supply Air Humidity (%)",
    "t_outdoor": "Outdoor Temperature (°C)",
    "t_saturation": "Saturation Temperature (°C)",
}

# Normal ranges (5th–95th percentile) based on NORMAL rows only

normal_df = df[df["anomaly_flag"] == 0]
range_info = normal_df[features].quantile([0.05, 0.95]).T
range_info.columns = ["p05", "p95"]
range_info = range_info.round(2)


def predict_hvac_status(
    power_kw,
    delta_supply_return,
    t_return,
    energy_kwh,
    t_supply,
    rh_outdoor,
    temp_error,
    rh_return,
    sp_return,
    rh_supply,
    t_outdoor,
    t_saturation,
):
    """
    Take a single set of HVAC readings and predict whether it is an anomaly or normal.
    Returns both the label and the model's probability for anomaly.
    """

    data = {
        "power_kw": [power_kw],
        "delta_supply_return": [delta_supply_return],
        "t_return": [t_return],
        "energy_kwh": [energy_kwh],
        "t_supply": [t_supply],
        "rh_outdoor": [rh_outdoor],
        "temp_error": [temp_error],
        "rh_return": [rh_return],
        "sp_return": [sp_return],
        "rh_supply": [rh_supply],
        "t_outdoor": [t_outdoor],
        "t_saturation": [t_saturation],
    }

    X_new = pd.DataFrame(data)[features]

    pred_class = rf.predict(X_new)[0]
    pred_proba = rf.predict_proba(X_new)[0][1]

    label = "ANOMALY" if pred_class == 1 else "NORMAL"

    print(f"Prediction: {label}")
    print(f"Anomaly probability: {pred_proba:.3f}")

    return label, pred_proba


def interactive_hvac_prediction():
    """
    Interactive tester that shows user-friendly feature names and normal ranges.
    Designed for recruiters, data scientists, or non-HVAC users.
    """

    print("Enter HVAC readings below.")
    print("Each input shows:")
    print("  - a human-friendly description")
    print("  - a typical NORMAL HVAC range (5th–95th percentile)")
    print("-------------------------------------------------------")

    row = {}

    for feat in features:
        display_name = friendly_names.get(feat, feat)
        low = range_info.loc[feat, "p05"]
        high = range_info.loc[feat, "p95"]

        while True:
            try:
                raw = input(f"{display_name}  (normal: {low} to {high}): ")
                value = float(raw)
                row[feat] = value
                break
            except ValueError:
                print("Please enter a numeric value.\n")

    print("\n--- Prediction Result ---")
    return predict_hvac_status(**row)
```

```python
# Run this cell to try custom values
interactive_hvac_prediction() 
```
